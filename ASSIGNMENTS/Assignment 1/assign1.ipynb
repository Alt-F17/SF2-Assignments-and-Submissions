6. The data in `data_df` contains noisy data.  In this step you are to clean the noisy data (or extract the clean data). Your cleaned data **for each run**
- must not include the data that is after once the filter hits the motion sensor
- must include the curved data that transitions into a drop in position.  You should only keep about 0.5 to 1 second of the roughly constant position values before the drop of position takes place.

7. Store your clean data into a new DataFrame called `clean_df`.

```python
# Your code for steps 6-9 go here:
KEEP = 0.75
clean_df = pd.DataFrame()
run_list = []
for i in range(1, 11):
    run_df = data_df[[f'Time (s) Run #{i}', f'Position (m) Run #{i}']].dropna()
    min_index = run_df[f'Position (m) Run #{i}'].idxmin()
    max_index = run_df[f'Position (m) Run #{i}'][:min_index].idxmax()
    max_time = run_df[f'Time (s) Run #{i}'][max_index]
    run_df = run_df[(run_df[f'Time (s) Run #{i}'] > max_time - KEEP) & (run_df.index <= min_index)]
    run_list.append(run_df)
clean_df = pd.concat(run_list, axis=1) # This literally made me cry... JF said it was okay to use this method to turn the list into a dataframe

clean_df.to_csv("clean_data_single_filter.csv", index=False)
```

10. Plot the data using `matplotlib` from all 10 runs as scatter plots on the same figure using your cleaned data.  
- Each plot should have a label **Run #$i$** for $1\leq i\leq10$
- The label on the x-axis should read **Time (s)**

11. Take the last 0.5 to 1 second of the x-values in your 10 runs, and draw a line of best fit on all your 10 graphs. 

```python
# Your code for steps 10-11 go here:

plt.figure(figsize=(12, 8))
REFERENCE_TIME = 1.5 # seconds
REFERENCE_POS = 0.4 # meters
FIT_LINE_INTERVAL = 1 # seconds
time_shifts = []

for i in range(1, 11):
    # For Loop 1: Finds reference point on each run
    pos_data = clean_df[f'Position (m) Run #{i}'].dropna()
    time_data = clean_df[f'Time (s) Run #{i}'].dropna()
    closest_idx = (pos_data - REFERENCE_POS).abs().idxmin() # Closest point to reference position
    actual_time = time_data[closest_idx] # Time at closest point
    time_shifts.append(REFERENCE_TIME - actual_time) # Shift the data according to that time

for i in range(1, 11):
    # For Loop 2: Shifts and plots the data
    pos_data = clean_df[f'Position (m) Run #{i}'].dropna()
    time_data = clean_df[f'Time (s) Run #{i}'].dropna()
    shifted_time = time_data + time_shifts[i-1]
    shifted_time = shifted_time[shifted_time >= 0]  # Remove negative times
    pos_data = pos_data[shifted_time.index]  # Filter out the corresponding positions
    plt.scatter(shifted_time, pos_data, label=f'Run #{i}', s=3)

for i in range(1, 11):
    # For Loop 3: Averages out the data for best fit line
    pos_data = clean_df[f'Position (m) Run #{i}'].dropna()
    time_data = clean_df[f'Time (s) Run #{i}'].dropna()
    shifted_time = time_data + time_shifts[i-1]
    end_time = shifted_time.max()
    start_time = end_time - 0.5  
    if i == 1:  # First run
        all_times = shifted_time[(shifted_time >= start_time) & (shifted_time <= end_time)]
        all_positions = pos_data[(shifted_time >= start_time) & (shifted_time <= end_time)]
    else:  # Add all the other runs
        all_times = np.concatenate([all_times, shifted_time[(shifted_time >= start_time) & (shifted_time <= end_time)]]) # JF said using concat was ok, specifically, to turn a list into an array
        all_positions = np.concatenate([all_positions, pos_data[(shifted_time >= start_time) & (shifted_time <= end_time)]])

# Best Fit Line Calculation & Plotting
coeffs = np.polyfit(all_times, all_positions, 1)
fit_line = np.poly1d(coeffs)
time_range = np.linspace(max(all_times)-FIT_LINE_INTERVAL, max(all_times), 100)
plt.plot(time_range, fit_line(time_range), 'r-', label='Best Fit Line', linewidth=2)

# Plotting Details
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs Time for All Runs (Single Filter) - Aligned at 1.5s')
plt.legend()
plt.grid(True)
plt.show()
```

5. The data in `data_df` contains noisy data.  In this step you are to clean the noisy data (or extract the clean data). Your cleaned data **for each run**
- must not include the data that is after once the filter hits the motion sensor
- must include only the bit where a drop in position starts.  You should not include any constant position values before the drop of positions takes place.    
Note that you should clean your data for each of the 10 runs and you should automate this process in your code with logical conditions, loops, etc.
6. Store your clean data into a new DataFrame called `clean_df`. 
7. Note that now that the values in the **Time** columns may have shifted due to the clean-up.  Update your `clean_df` to rescale all the **Time** columns to start the time from 0.0 in increments of 0.02. 
8. Write this to a new .csv file called `"clean_data_multiple_filter.csv"`. 

```python
# # Your code for steps 5 and 6 go here:
def clean(df: pd.DataFrame) -> pd.DataFrame:
    cleaned = df[: df.loc[df[df.columns[1]] == df[df.columns[1]].min(), df.columns[1]].index[0] + 1]

    possible_bumps = 0
    start = 0
    for idx, pos in enumerate(reversed(cleaned[df.columns[1]])):
        last_position = cleaned[df.columns[1]].iloc[len(cleaned) - idx] if idx != 0 else 0
        change = pos - last_position

        if change < 0.02:
            possible_bumps += 1 
        else:
            possible_bumps = 0 
        
        if possible_bumps >= 5 and start == 0:
            start = len(cleaned) - idx + 2

    return cleaned[start:]

clean_df = pd.DataFrame()
for i in range(1, 11):
    run_df = data_df[[f'Time (s) Run #{i}', f'Position (m) Run #{i}']].dropna()
    cleaned_run = clean(run_df)
    if i == 1:
        clean_df = cleaned_run
    else:
        clean_df = clean_df.join(cleaned_run, how='outer')

clean_df.to_csv("clean_data_multiple_filter.csv", index=False)
clean_df

```
