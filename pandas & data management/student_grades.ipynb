```python
# grades.dropna().plot(kind='scatter', x='Hours', y='Scores')

# OR

plt.scatter(grades['Hours'], grades['Scores'], s=5)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours vs Scores')
plt.show()
```

```python
# Getting Minimum and Maximum values:

print(grades['Scores'].min())
print(grades['Scores'].max())

# OR

grades['Scores'].describe()
```

```python
# Changing value type in a column:

# grades['Scores'] = grades['Scores'].astype(str)

new_grades = grades.set_index('Hours').sort_values('Hours')
print(new_grades)

# Pulling info from that first row

print(new_grades.loc[1.1])

# Pulling info from a slice of the DataFrame

print(new_grades.loc[1.1:5.5]) # only works if the DataFrame is sorted

```

```python
# Renaming things (columns, indexes, etc.)

new_grades2 = grades.rename(columns={'Hours': 'Hours Studied'})
print(new_grades2)

new_grades2 = new_grades.rename_axis('Hours Studied')
print(new_grades2)

```

# Loc vs iLoc

```python
# Get the index of rows where hours is equal to 1.1 using loc & iloc

list = grades.index[grades['Hours']==1.1].tolist()
print(list)

# OR

list = grades.index[grades['Hours']==2.5].tolist()
print(list)

index = list[1]

grades.loc[index]

# iloc

print(grades.iloc[index])
print(grades.iloc[index:index+6])
```

# at, iat && slicing

```python
new_grades.loc[1.1:2.5, ['MidTerm', 'FinalExam', 'Scores']] # This will give you the values of the columns 'MidTerm', 'FinalExam', and 'Scores' 
                                                          # for the rows where the 'Hours' column is between 1.1 and 2.0

grades.iloc[[0,5], 0:4] # This will give you the values of the first 5 columns for the rows with index 0 and 5

new_grades.at[1.1, 'FinalExam'] # This will give you the value of the 'FinalExam' column for the row with index 1.1
grades.iat[2, 4] # This will give you the value of the 4th column for the row with index 0

```
